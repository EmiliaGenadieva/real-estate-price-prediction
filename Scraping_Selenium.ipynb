```python
from time import sleep
import bs4
from selenium import webdriver
from selenium.webdriver import Firefox
import pandas as pd
import threading
import numpy as np
links = []
living_area = []
living_room_surface = []
kitchen_type = []

with open('links.txt') as f:
    links = f.read().split(",") 
    for var in links:   
    
        try:
            drive = webdriver.Firefox()
            #thread = threading.Thread(target=drive.get(var))
            thread = threading.Thread(target=drive.get(url='https://www.immoweb.be/en/classified/house/for-sale/la-louviere/7100/10301018'))
            thread.start()
            sleep(3)
            soup = bs4.BeautifulSoup(drive.page_source, "html.parser")
            jobs_elements = soup.find_all("tr", class_= "classified-table__row")
            data = []
            for jobs_element in jobs_elements:
                try:
                    table_data = jobs_element.find("td", class_="classified-table__data").text.strip('\n\n ')
                except AttributeError:
                    table_data = 'no data'
                try:
                    table_header = jobs_element.find("th", class_="classified-table__header").text.strip('\n\n ')
                except AttributeError:
                    table_header = 'no data'
                    
                item = {'data':table_data,
                        'header': table_header}   
                data.append(item)
            df = pd.DataFrame(data)
                
            if 'Living area' in df.iloc[7][1]:
                living_area.append(df.iloc[7][0])
            else:
                living_area.append(np.NaN)
            if 'Living room surface' in df.iloc[8][1]:
                living_room_surface.append(df.iloc[8][0])
            else:
                living_room_surface.append(np.NaN)
            if 'Kitchen type' in df.iloc[10][1]:
                kitchen_type.append(df.iloc[10][0])
            else:
                    kitchen_type.append(np.NaN)
                
            my_dict = {'Living area': living_area, 'Living room surface': living_room_surface, 'Kitchen type': kitchen_type}
            thread.join()
            drive.close()
        except TimeoutError:
            print('This is TimeoutError')
            drive.close()
        except:
            sleep(3)
            continue
    
print(pd.DataFrame(my_dict))
```
